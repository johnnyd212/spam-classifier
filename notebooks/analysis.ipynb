## Pre-processing

### Loading the data

Our CSV file is unlabeled, so we create our column names following the description in spambase.DOCUMENTATION:

```python
word_freq_names = [f'word_freq_{i}' for i in range(1, 49)]
char_freq_names = [f'char_freq_{i}' for i in range(1, 7)]
cap_run_features = ['capital_run_length_' + s for s in ['average', 'longest', 'total']]

col_names = (
    word_freq_names +
    char_freq_names +
    cap_run_features +
    ['spam']
)
```

```python
df = pd.read_csv('../data/spambase.data', header=None, names=col_names)
df.head()
```

### Feature normalization

We can see that the ranges of the values for our integral features are quite large relative to our word and char frequencies:

```python
df[cap_run_features].describe()
```

I tried to train the regressor without normalizing at first, but the sigmoid function saturated, and we were getting numerical issues of taking logs of 0 in our cross entropy loss as a consequence. To address this, we min-max normalize all of our features.

```python
minima = df.min()
maxima = df.max()

df = (df - minima) / (maxima - minima) 
```

```python
df.describe()
```

### Random Seed

For reproducibility, we'll set the seed for the random generators we use (e.g. for train-test spliiting, stochastic gradient descent, etc...).

```python
rng_seed = 2026
rng = np.random.default_rng(seed=rng_seed)
```

### Train-test split

We'll implement a train-test split of 5-95, that is, we use 5% of the data for training, and hold out the remaining 95% for testing. (In the interest of implementing our classifier from 'scratch', we don't use train_test_split from scikit-learn.) We choose relatively small training sample from our data to deliberately induce overfitting, thereby making the effects of regularization clearer later.

```python
train_ratio = 0.05
N = len(df)

# Train-test split
df_perm = df.iloc[rng.permutation(N)]
split_index = int(N * train_ratio)
df_train = df_perm.iloc[:split_index]
df_test = df_perm.iloc[split_index:]

X_train = df_train.drop(columns=['spam'])
y_train = df_train['spam']
X_test = df_test.drop(columns=['spam'])
y_test = df_test['spam']
```

## Logistic Regression with SGD

```python
from regressor import LogisticRegressor
```

We'll initialize two logistic regressors, one with no regularization and the other with regularization strength 0.001, to compare their loss curves.

Additionally, we initialize a third logistic regressor with \textit{momentum}. With momentum rate parameter $\beta$ (called momentum_rate in our code), our $t$th update step becomes

$$ M^{(t)} = \beta M^{(t-1)} + (1 - \beta)\nabla_w J(w)$$
$$ w^{(t)} = w - \alpha * M^{(t)}$$

We'll initialize our hyperparameters as follows:

```python
reg_str_init = 0.01
batch_size_init = 1
learning_rate_init = 0.001
num_epochs_init = 100
momentum_rate_init = 0.9995
```

Additionally, to observe what happens when our momentum gets too large, will initialize another regressor with $\beta$ set to $0.9999$.

```python
momentum_rate_excess = 0.9999
```

```python
model_basic = LogisticRegressor(
    batch_size=batch_size_init,
    learning_rate=learning_rate_init,
    num_epochs=num_epochs_init,
    seed=2026
)

model_reg = LogisticRegressor(
    batch_size=batch_size_init,
    learning_rate=learning_rate_init,
    num_epochs=num_epochs_init,
    regularization_strength=reg_str_init, 
    seed=2026
)

model_mom = LogisticRegressor(
    batch_size=batch_size_init,
    learning_rate=learning_rate_init,
    num_epochs=num_epochs_init,
    momentum_rate = momentum_rate_init, 
    seed=2026
)

model_mom_excess = LogisticRegressor(
    batch_size=batch_size_init,
    learning_rate=learning_rate_init,
    num_epochs=num_epochs_init,
    momentum_rate = momentum_rate_excess, 
    seed=2026
)  
```

```python
learning_curve_basic = model_basic.fit(
    X_train.to_numpy(), 
    y_train.to_numpy(), 
    learning_curve=True
)

learning_curve_reg = model_reg.fit(
    X_train.to_numpy(), 
    y_train.to_numpy(), 
    learning_curve=True
)               

learning_curve_mom = model_mom.fit(
    X_train.to_numpy(), 
    y_train.to_numpy(), 
    learning_curve=True
)       

learning_curve_mom_excess = model_mom_excess.fit(
    X_train.to_numpy(), 
    y_train.to_numpy(), 
    learning_curve=True
)     
```

```python
plt.plot(np.arange(1, model_basic.num_epochs + 1), 
            learning_curve_basic, 
            'r--', label='\u03BB = 0, \u03B2 = 0')
plt.plot(np.arange(1, model_reg.num_epochs + 1),
            learning_curve_reg,
            'b--', label=f'\u03BB = {reg_str_init}, \u03B2 = 0')
plt.plot(np.arange(1, model_mom.num_epochs + 1),
            learning_curve_mom,
            'g--', label=f'\u03BB = 0, \u03B2 = {momentum_rate_init}')
plt.plot(np.arange(1, model_mom_excess.num_epochs + 1),
            learning_curve_mom_excess,
            'y--', label=f'\u03BB = 0, \u03B2 = {momentum_rate_excess}')

plt.xlabel("Epoch") 
plt.ylabel("Cost")
plt.title("Learning Curves")
plt.legend()
```

The plot shows the learning curves for our model initialized with a few fixed parameters. The regularization strength of $\lambda=0.01$ elicits a modest improvement from the base model (i.e. the model without regularization or momentum), however it is important to remember that with regularization *we have changed the loss function*, and therefore the global minima we reach in the new loss landscape may actually differ substantially from the global minimum of the original function. The consequences of high bias become obvious when we investigate validation error.

On the other hand, applying momentum with $\beta = 0.9995$ yields a noticeably faster convergence to the optimum compared to the models without momentum. Increasing $\beta$ to $0.9999$ speeds up convergence slightly further at first, however it comes at the expense of our momentum becoming so large that we "overshoot" the optimum and our loss actually begins to increase again before eventually converging again.

For all four models, we observe good stability. Loss decreases smoothly for the majority of training, with a slight uptick in loss becoming noticeable around epoch 80 as numerical instability starts to compound.
